# มาติดเกาะกันเถอะทุกคนนนน

```python
Individual2 = "A floating seat or cushion เพราะว่าต้องใช้ตอนลอยเข้าฝั่งให้ได้ค่ะ จะไม่ลอยคออยู่กลางทะเลแน่นอน"
for i in Individual2:
    old = ord(i)+1
    x += chr(old)
print("old data: ",Individual2)
print("new data: ",x)

print("\n------------------------------------------------------------\n")

Group2 = "A can of petrol เพราะว่าใช้เป็นเชื่อเพลิงในการจุดไฟ,และถ้าเรือรั่วก็เอามาใช้เป็นทุ่นลอยน้"
for i in Group2:
    old = ord(i)+1
    x += chr(old)
print("old data: ",Group2)
print("new data: ",x)
```
